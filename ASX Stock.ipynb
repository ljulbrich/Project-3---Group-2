```python
import pandas as pd
from pathlib import Path
import concurrent.futures
import yfinance as yf
import csv

```

```python
file_path = Path("Resources/30-June-2023 ASX Stock Data.csv")
```

```python
asx_df = pd.read_csv(file_path)
```

```python
asx_df
```

```python
asx_df['Product Description Abbrev'] = asx_df['Product Description Abbrev'].str.replace('ORDINARY            ', 'ORDINARY')

asx_df
```

```python
filtered_df = asx_df[(asx_df['Security Group Code'] == 'EQY') & (asx_df['Product Description Abbrev'] == 'ORDINARY')]

filtered_df
```

```python
filtered_df['ASX Code'] = filtered_df['ASX Code'].str.strip()

filtered_df
```

```python
filtered_df['ASX Code'] = filtered_df['ASX Code'].astype(str) + ".AX"

filtered_df
```

```python
asx_code_list = filtered_df['ASX Code'].tolist()

asx_code_list
```

```python
import concurrent.futures
import yfinance as yf
import csv

def get_stock_info(asx_code):
    try:
        ticker = yf.Ticker(asx_code)
        name = ticker.info.get('longName', 'N/A')
        industry = ticker.info.get('industry', 'N/A')
        sector = ticker.info.get('sector', 'N/A')
        market_cap = ticker.info.get('marketCap', 'N/A')
        employee_num = ticker.info.get('fullTimeEmployees', 'N/A')
        revenue = ticker.info.get('totalRevenue', 'N/A')
        price = ticker.info.get('currentPrice', 'N/A')
        volume = ticker.info.get('averageVolume', 'N/A')
        pe_ratio = ticker.info.get('trailingPE', 'N/A')
        eps = ticker.info.get('trailingEps', 'N/A')
        roe = ticker.info.get('returnOnEquity', 'N/A')
        
        
        # Check if market_cap data is available
        if market_cap != 'N/A':
            return {'ASX Code': asx_code, 'Name': name, 'Industry': industry,
                    'Sector': sector, 'Market Cap': market_cap,
                    'Employee Number': employee_num, 'Total Revenue': revenue,
                    'Price': price, 'Volume': volume, 'P/E Ratio': pe_ratio,
                    'Earning Per Share': eps, 'Return On Equity': roe}
        else:
            return {'ASX Code': asx_code, 'Market Cap': 'Not available'}
    except Exception as e:
        return None  # Return None if an error occurs

# Define a ThreadPoolExecutor with a max_workers parameter for parallel processing
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    # Submit tasks for each ASX code to fetch stock info
    futures = [executor.submit(get_stock_info, asx_code) for asx_code in asx_code_list]
    
    # Open a CSV file in write mode and create a CSV writer object
    with open('ASX Output/stock_info_results.csv', 'w', newline='') as csvfile:
        fieldnames = ['ASX Code', 'Name', 'Industry', 'Sector', 'Market Cap',
                      'Employee Number', 'Total Revenue', 'Price', 'Volume',
                      'P/E Ratio', 'Earning Per Share', 'Return On Equity']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        # Write the header row to the CSV file
        writer.writeheader()
        
        # Process the futures and write results to CSV file
        for future in concurrent.futures.as_completed(futures):
            result = future.result()
            writer.writerow(result)
```

```python
output_file = 'ASX Output/stock_info_results.csv'  
asx_stock_df = pd.read_csv(output_file)

asx_stock_df
```
